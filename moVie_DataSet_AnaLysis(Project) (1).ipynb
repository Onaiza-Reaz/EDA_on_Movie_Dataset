# Classify Movies Based On Rating ["EXCELLENT","GOOD","AVERAGE"]

```python
df.columns
```

```python
def Rating(rating):
  if rating >= 7.0:
    return " Excellent"
  elif rating >=6.0:
    return "Good"
  else:
    return "Average"
```

```python
df["Rating_Remarks"] = df["Rating"].apply(Rating)
```

# Find Unique Values From Genre

```python
df["Genre"]
```

```python
two_D =[]
for  value in df["Genre"]:
  two_D.append(value.split(","))
two_D
```

```python
one_D=[]
for item in two_D:
  for item1 in item:
    one_D.append(item1)
```

```python
one_D
```

```python
uni_list=[]
for item  in one_D:
  if item not in uni_list:
    uni_list.append(item)
```
